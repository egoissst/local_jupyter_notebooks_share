```python
import pandas as pd
```

```python
file_loc = "/Users/yash.dalmia/Downloads/rejected_count_jan_rejected_count_jan.csv"
```

```python
df = pd.read_csv(file_loc, names=['C_D', 'C_APP_RES', 'Count'], header=0)
```

```python
import time
```

```python
def get_datetime_string_ist(epoch_millis):
#     return time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(int(epoch_millis)/1000))
    return time.strftime('%Y-%m-%d', time.localtime(int(epoch_millis)/1000))
```

```python
def convert_dataframe_unstack(df):
    index_list = [(row['C_D'], row['C_APP_RES']) for index, row in df.iterrows()]
    index = pd.MultiIndex.from_tuples(index_list)
    series = pd.Series(df['Count'].tolist(), index = index)
    df_converted = series.unstack(level=1)
    return df_converted
```

```python
df_converted = convert_dataframe_unstack(df)
```

```python
df_converted.fillna(0, inplace=True)
```

```python
df_converted
```

```python
df_converted = df_converted.astype('int32')
```

```python
df_converted['Date'] = df_converted.index
df_converted['Date'] = df_converted['Date'].apply(lambda x: get_datetime_string_ist(x))
```

```python
df_converted.to_csv('/Users/yash.dalmia/Downloads/converted_df.csv')
```
