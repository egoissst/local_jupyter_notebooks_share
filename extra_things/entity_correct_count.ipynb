```python
def get_entity_correct_api(appkey, msid, to_take_action):
    url = "http://192.168.43.239/mytimes/correctEntityNReplyCount?appkey={}&msid={}&toTakeAction={}".format(appkey, msid, to_take_action)
    return url
```

```python
to_take_action = 'true'
cnt = 0
for index, x in enumerate(list_msid):
    fin_url = get_entity_correct_api(x[1], x[0], to_take_action)
    res_ = requests.get(fin_url)
    print(index+1, x[1], x[0], res_.text)
    if(res_.text == 'true'):
        cnt = cnt + 1

net_length = len(list_msid)
frac = cnt*1.0/(len(list_msid))

print('cnt', cnt, sep=" : ")
print('len', net_length, sep=" : ")
print('fraction', frac, sep=" : ")
```
